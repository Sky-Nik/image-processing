# Вирівнювання гістограми
----
_Ця стаття перекладена мною з англійської Вікіпедії._

**Вирівнювання гістограми** (ВГ, англ. histogram equalization, HE) це метод у [обробці зображень](https://uk.wikipedia.org/wiki/%D0%9E%D0%B1%D1%80%D0%BE%D0%B1%D0%BA%D0%B0_%D0%B7%D0%BE%D0%B1%D1%80%D0%B0%D0%B6%D0%B5%D0%BD%D1%8C), який покращує [контраст](https://uk.wikipedia.org/wiki/%D0%9A%D0%BE%D0%BD%D1%82%D1%80%D0%B0%D1%81%D1%82) з використанням гістограми [зображення](https://uk.wikipedia.org/wiki/%D0%97%D0%BE%D0%B1%D1%80%D0%B0%D0%B6%D0%B5%D0%BD%D0%BD%D1%8F).

Гістограма з яскраво вираженим піком трансформується таким чином, щоб пік заповнив більшу частину вісі $Ox$:

![Гістограми зображення до і після перетворення](https://upload.wikimedia.org/wikipedia/commons/c/ca/Histogrammeinebnung.png)

## Огляд
----
Цей метод зазвичай збільшує глобальний [контраст](https://uk.wikipedia.org/wiki/%D0%9A%D0%BE%D0%BD%D1%82%D1%80%D0%B0%D1%81%D1%82) зображень, особливо коли корисні [дані](https://uk.wikipedia.org/wiki/%D0%94%D0%B0%D0%BD%D1%96) у зображенні знаходяться у вузькому діапазоні інтенсивностей. Шляхом вирівнювання ці [інтенсивності](https://uk.wikipedia.org/wiki/%D0%A1%D0%B8%D0%BB%D0%B0_%D1%81%D0%B2%D1%96%D1%82%D0%BB%D0%B0) краще розподіляються гістограмою. Це дозволяє областям з низьким локальним контрастом набути вищого контрасту. Вирівнювання гістограми досягає цього шляхом перерозподілу найбільш частих значень інтенсивності.

Метод корисний для зображень у яких як тло так і передній план або обидва світлі, або обидва темні. Зокрема, метод може призвести до кращого відображення структури [кісток](https://uk.wikipedia.org/wiki/%D0%9A%D1%96%D1%81%D1%82%D0%BA%D0%B0) у [рентгенівських](https://uk.wikipedia.org/wiki/%D0%A0%D0%B5%D0%BD%D1%82%D0%B3%D0%B5%D0%BD%D1%96%D0%B2%D1%81%D1%8C%D0%BA%D0%B5_%D0%B2%D0%B8%D0%BF%D1%80%D0%BE%D0%BC%D1%96%D0%BD%D1%8E%D0%B2%D0%B0%D0%BD%D0%BD%D1%8F) знімках, а також до кращої видимості деталей на [фотокартках](https://uk.wikipedia.org/wiki/%D0%A4%D0%BE%D1%82%D0%BE%D0%B3%D1%80%D0%B0%D1%84%D1%96%D1%87%D0%BD%D0%B8%D0%B9_%D1%82%D0%B2%D1%96%D1%80), знятих із недостатньою, чи надмірною експозицією. Ключовими перевагами методу є доволі прямолінійна реалізація, а також можливість [зворотнього](https://uk.wikipedia.org/wiki/%D0%9E%D0%B1%D0%B5%D1%80%D0%BD%D0%B5%D0%BD%D0%B8%D0%B9_%D0%B5%D0%BB%D0%B5%D0%BC%D0%B5%D0%BD%D1%82) перетворення. Тобто, в теорії, якщо знати <a href="https://uk.wikipedia.org/wiki/%D0%A4%D1%83%D0%BD%D0%BA%D1%86%D1%96%D1%8F_(%D0%BC%D0%B0%D1%82%D0%B5%D0%BC%D0%B0%D1%82%D0%B8%D0%BA%D0%B0)">функцію</a> вирівнювання гістограми, то з результуючої гістограми можна поновити оригінальну гістограму. Обчислювальна складність алгоритму невисока. Недоліком методу є його безрозбіливість: він може збільшити контраст [шуму](https://uk.wikipedia.org/wiki/%D0%97%D0%B0%D0%B2%D0%B0%D0%B4%D0%B8), таким чином приглушуючи корисний [сигнал](https://uk.wikipedia.org/wiki/%D0%A1%D0%B8%D0%B3%D0%BD%D0%B0%D0%BB).

У наукових зображення, де просторова кореляція важливіша за інтенсивність сигналу (наприклад, розділення фрагментів [ДНК](https://uk.wikipedia.org/wiki/%D0%94%D0%B5%D0%B7%D0%BE%D0%BA%D1%81%D0%B8%D1%80%D0%B8%D0%B1%D0%BE%D0%BD%D1%83%D0%BA%D0%BB%D0%B5%D1%97%D0%BD%D0%BE%D0%B2%D0%B0_%D0%BA%D0%B8%D1%81%D0%BB%D0%BE%D1%82%D0%B0) з дискретизованою довжиною), мале [співвідношення сигнал/шум](https://uk.wikipedia.org/wiki/%D0%A1%D0%BF%D1%96%D0%B2%D0%B2%D1%96%D0%B4%D0%BD%D0%BE%D1%88%D0%B5%D0%BD%D0%BD%D1%8F_%D1%81%D0%B8%D0%B3%D0%BD%D0%B0%D0%BB/%D1%88%D1%83%D0%BC) зазвичай перешкоджає візуальному виявленню.

Вирівнювання гістограми часто створює нереалістичні ефекти на фотографіях; однак це дуже корисно для таких наукових зображень, як [теплові](https://uk.wikipedia.org/wiki/%D0%A2%D0%B5%D0%BF%D0%BB%D0%BE%D0%B1%D0%B0%D1%87%D0%B5%D0%BD%D0%BD%D1%8F), [супутникові](https://uk.wikipedia.org/wiki/%D0%A8%D1%82%D1%83%D1%87%D0%BD%D0%B8%D0%B9_%D1%81%D1%83%D0%BF%D1%83%D1%82%D0%BD%D0%B8%D0%BA) або [рентгенівські](https://uk.wikipedia.org/wiki/%D0%A0%D0%B5%D0%BD%D1%82%D0%B3%D0%B5%D0%BD%D1%96%D0%B2%D1%81%D1%8C%D0%BA%D0%B5_%D0%B2%D0%B8%D0%BF%D1%80%D0%BE%D0%BC%D1%96%D0%BD%D1%8E%D0%B2%D0%B0%D0%BD%D0%BD%D1%8F) знімки. Часто це зображення того ж класу, до якого доцільно застосовувати [несправжні кольори](https://en.wikipedia.org/wiki/False_color). Також вирівнювання гістограми може спричинити небажані ефекти (наприклад, видимий градієнт зображення) при застосуванні до зображень з низькою [глибиною кольору](https://uk.wikipedia.org/wiki/%D0%93%D0%BB%D0%B8%D0%B1%D0%B8%D0%BD%D0%B0_%D0%BA%D0%BE%D0%BB%D1%8C%D0%BE%D1%80%D1%83). Наприклад, якщо застосувати до 8-бітного зображення, що відображається за допомогою 8-бітної палітри сірого кольору, це ще більше зменшить глибину кольору (кількість унікальних відтінків сірого) зображення. Вирівнювання гістограми працюватиме найкраще при застосуванні до зображень із значно більшою [глибиною кольору](https://uk.wikipedia.org/wiki/%D0%93%D0%BB%D0%B8%D0%B1%D0%B8%D0%BD%D0%B0_%D0%BA%D0%BE%D0%BB%D1%8C%D0%BE%D1%80%D1%83), ніж розмір палітри, як-от [неперервні](https://uk.wikipedia.org/wiki/%D0%9D%D0%B5%D0%BF%D0%B5%D1%80%D0%B5%D1%80%D0%B2%D0%BD%D0%B0_%D1%84%D1%83%D0%BD%D0%BA%D1%86%D1%96%D1%8F) дані або 16-бітні зображення сірого масштабу.

Є два способи реалізації вирівнювання гістограми. Перший це зміна власне зображення, а другий --- зміна палітри. Операція вирівнювання може бути виражена як $P(M(I))$, де $I$ --- вихідне зображення, $M$ --- оператор вирівнювання гістограми, а $P$ --- палітра. Якщо ж ми визначимо нову палітру як $P' = P(M)$ і залишимо зображення $I$ незмінним, то гістограми реалізується як зміна палітри. З іншого боку, якщо палітра $P$ залишається незмінною а зображення модифікується як $I' = M(I)$, то реалізація відбувається за допомогою зміни зображення. У більшості випадків зміна палітри краще, оскільки вона зберігає вихідні дані (а отже може бути обернена за потреби).

Модифікації цього методу використовують кілька гістограм, так званих субгістограмам, щоб підкреслити локальний, а не загальний контраст. Прикладом такого методу є адаптивне вирівнювання гістограми, контрастно-обмежене адаптивне вирівнювання гістограми (CLAHE), багато-пікове вирівнювання гістограми (англ. MPHE) і бета-оптимізоване вирівнювання гістограми широкого призначення (англ. MBOBHE). Метою цих методів, особливо MBOBHE, є поліпшення контрасту без створення артефактів накшталт зсуву середньої яскравості і втрати деталей.<sup>[1]</sup>

Трансформація сигналу, нібито еквівалентна вирівнюванню гістограми, також відбувається в [біологічних нейронних мережах](https://uk.wikipedia.org/wiki/%D0%91%D1%96%D0%BE%D0%BB%D0%BE%D0%B3%D1%96%D1%87%D0%BD%D0%B0_%D0%BD%D0%B5%D0%B9%D1%80%D0%BE%D0%BD%D0%BD%D0%B0_%D0%BC%D0%B5%D1%80%D0%B5%D0%B6%D0%B0), щоб максимізувати швидкість випуску нейрона як функцію вхідної статистики. Це було доведено, зокрема, на [сітківці](https://uk.wikipedia.org/wiki/%D0%A1%D1%96%D1%82%D0%BA%D1%96%D0%B2%D0%BA%D0%B0) очей [мух](https://uk.wikipedia.org/wiki/%D0%94%D0%B2%D0%BE%D0%BA%D1%80%D0%B8%D0%BB%D1%96).<sup>[2]</sup>

Вирівнювання гістограми є конкретним випадком більш загального класу методів перекомпонування гістограм. Ці методи спрямовані на коригування зображення, щоб полегшити аналіз або покращити якість візуалізації (наприклад, [retinex](https://en.wikipedia.org/wiki/Color_constancy#Retinex_theory)).


### Зворотня проекція

Зворотня проекція гістограмованого зображення --- це повторне застосування модифікованої гістограми до вихідного зображення, що функціонує як хеш-таблиця значень яскравості пікселів.

Для кожної групи пікселів, отриманих з одного і того ж положення по усім одноканальним зображенням, функція переводить номер кошика гістограми у цільове зображення, де номер кошика визначається значеннями інтенсивностей пікселів цієї групи. З точки зору статистики, значення кожного пікселя вихідного зображення характеризує ймовірність того, що відповідна група вхідних пікселів належить об'єкту, гістограма якого використовується.<sup>[3]</sup>

## Реалізація
----
Розглнемо дискретне чорно-біле зображення $x$ і нехай $n_i$ --- кількість разів, які у зображення зустрічається інтенсивність $i$. Ймовірність обрати піксель інтенсивності $i$ вибираючи навмання рівна

$$
p_x(i) = p(x = i) = \frac{n_i}{n}, \quad 0 \le i < L,
$$

де $L$ --- максимально можлива інтенсивність (зазвичай $256$), $n$ --- загальна кількість пікселів у зображенні, а $p_x(i)$ --- по суті нормоване на $[0, 1]$ значення гістограми у точці, що відповідає інтенсивності $i$.

Визначимо також [функцію розподілу](https://uk.wikipedia.org/wiki/%D0%A4%D1%83%D0%BD%D0%BA%D1%86%D1%96%D1%8F_%D1%80%D0%BE%D0%B7%D0%BF%D0%BE%D0%B4%D1%96%D0%BB%D1%83_%D0%B9%D0%BC%D0%BE%D0%B2%D1%96%D1%80%D0%BD%D0%BE%D1%81%D1%82%D0%B5%D0%B9) $p_x$ як

$$
\text{cdf}_x(i) = \sum_{j = 0}^i p_x(x = j),
$$

яка також називається кумулятивною нормалізованю гістограмою зображення.

Ми хочемо отримати перетворення вигляду $y = T(x)$ щоб отримати нове зображення $y$, із вирівняною гістограмою. Таке зображення повинне мати майже лінійну функцію розподілу, тобто

$$
\text{cdf}_y (i) = i K
$$

для певної сталої $K$. Властивості функції розподілу дозволяють нам одазу виписати потрібне нам перетворення (а саме, обернена функція роподілу); воно визначається формулою

$$
\text{cdf}_y (y') = \text{cdf}_y (T(k)) = \text{cdf}_x (k)
$$

де $k$ пробігає значення з проміжку $[0,L]$). Зауважте, що $T$ переводить значення аргументу у проміжок $[0,1]$, оскільки ми використали нормалізовану гістограму $x$. Аби повернути значення назад на початковий проміжок необхідно застосувати ще таке наступне просте лінійне перетворення до результату попереднього:

$$
y' = y \cdot (\max x - \min x) + \min x
$$

Детальніше виведення доступне за [посиланням](https://www.math.uci.edu/icamp/courses/math77c/demos/hist_eq.pdf).

## Кольорові зображення
----
Вище описане вирівнювання гістограми чорно-білого зображення. Однак, воно (вирівнювання гістограми) також застосовне до кольоровних зображень, шляхом поканального застосування того ж методу окремо до червоної, зеленої і синьої компонент [RGB](https://uk.wikipedia.org/wiki/RGB)-хображення. Щоправда, роздільне застосування до кожного каналу може призвести до драматичних змін у [балансі білого](https://uk.wikipedia.org/wiki/%D0%91%D0%B0%D0%BB%D0%B0%D0%BD%D1%81_%D0%B1%D1%96%D0%BB%D0%BE%D0%B3%D0%BE_%D0%BA%D0%BE%D0%BB%D1%8C%D0%BE%D1%80%D1%83) кольору зображення, адже відносні розподіли кольорів змінюються в результаті застосування алгоритму. Втім, якщо спершу конвертувати зображення до іншого колірного простору, наприклад, [HSL](https://uk.wikipedia.org/wiki/HSL), то алгоритм може бути застосований тільки до складової [яскравості](https://uk.wikipedia.org/wiki/%D0%AF%D1%81%D0%BA%D1%80%D0%B0%D0%B2%D1%96%D1%81%D1%82%D1%8C), без змін у віддтінку і насиченості зображення.<sup>[4]</sup>

Таокж є кілька технік вирівнювання гістограм у тривимірному колірному просторі.<sup>[5]</sup> Щоправда, вони призводять до "засвічування", тобто до ситуації, коли ймовірність світліших пікселів вища за ймовірність темніших.<sup>[6]</sup> Han та ін. запропонували виористовувати нову функцію розподілу, що визначається у площині сталої світлості (анлг. iso-luminance plane), що призвело до рівномірного розподілу яскравості.<sup>[7]</sup>

## Приклади

-----

For consistency with statistical usage, "CDF" (i.e. Cumulative distribution function) should be replaced by "cumulative histogram", especially since the article links to cumulative distribution function which is derived by dividing values in the cumulative histogram by the overall amount of pixels. The equalized CDF is defined in terms of rank as $\text{rank}/\text{pixelcount}$.

# Адаптивне вирівнювання гістограми
----
_Ця стаття перекладена мною з англійської Вікіпедії._

**Адаптивне вирівнювання гістограми** (АВГ, англ. adaprive histogram equalization, AHE) --- техніка комп'ютерної [обробки зображень](https://uk.wikipedia.org/wiki/%D0%9E%D0%B1%D1%80%D0%BE%D0%B1%D0%BA%D0%B0_%D0%B7%D0%BE%D0%B1%D1%80%D0%B0%D0%B6%D0%B5%D0%BD%D1%8C), яка застосовується для покращення [контрастності](https://uk.wikipedia.org/wiki/%D0%9A%D0%BE%D0%BD%D1%82%D1%80%D0%B0%D1%81%D1%82) зображень. Вона відрізняється від звичайного вирівнювання гістограми у тому, що адаптивний метод обчислює кілька [гістограм](https://uk.wikipedia.org/wiki/%D0%93%D1%96%D1%81%D1%82%D0%BE%D0%B3%D1%80%D0%B0%D0%BC%D0%B0), кожна з яких відповідає окремій частині зображення, а потім використовує їх для перерозподілу світлості по зображенню. Як наслідок, вона краще підходить до покращення локальної контрастності, і краще виділяє краї в усіх областях зображення.

Щоправда, на практиці АВГ має тенденцію до збільшення [шуму](https://uk.wikipedia.org/wiki/%D0%97%D0%B0%D0%B2%D0%B0%D0%B4%D0%B8) у відносно монотонних областях зображення. Варіант адаптивного вирівнювання гістограми, який називається контрастно-обмежене адаптивне вирівнювання гістограми (КОАВГ, англ. CLAHE) попереджує це шляхом обмеження збільшення контрасту.

## Мотивація й опис методу

-----

Звичайне вирівнювання гістограми використовує одне і те ж перетворення, отримане з початкової гістограми зображення, для перетворення усіх пікселей. Це гарно працює, коли розподіл інтенсивностей пікселей схожий по всьому зображенні. Втім, якщо зображення містить суттєво світліші чи суттєво темніші ніж більшість зображення області, то контраст у цих областях не буде суттєво покращений.

Адаптивне вирівнювання гістограми (АВГ) покращує ситуацію трансформуючи кожен піксель за допомогою еретворення, обчисленого на основі до цього пікселя сусідніх областей. Воно спрешу було розроблене для застосування у дисплеях літаків-винищувачів.<sup>[1], цитовано у [2]</sup> У своїй найпростішій формі, кожен піксель трансформується на підставі гістограми у невеликому квадраті, що його оточує, як на зобрженні нижче. Виведення перетворювальних функцій з гістограми залишається таким же, як і для звичайного вирівнювання гістограми: перетворювальна функція пропорційні [функції розподілу](https://uk.wikipedia.org/wiki/%D0%A4%D1%83%D0%BD%D0%BA%D1%86%D1%96%D1%8F_%D1%80%D0%BE%D0%B7%D0%BF%D0%BE%D0%B4%D1%96%D0%BB%D1%83_%D0%B9%D0%BC%D0%BE%D0%B2%D1%96%D1%80%D0%BD%D0%BE%D1%81%D1%82%D0%B5%D0%B9) інтенсивностей пікселів у околі.

![AHE-neighbourhoods](https://upload.wikimedia.org/wikipedia/commons/thumb/8/84/AHE-neighbourhoods.svg/432px-AHE-neighbourhoods.svg.png)

Пікселі поблизу краю зображення необхідно обробляти спеціальним чином, адже їхні околи не лежать цілком всередині зображення. Це застосовно, наприклад, до пікселів, що лежаьть лівіше чи вище за синій пікселі на зображенні вище. Цю проблему можна вирішити якщо розширити зображення шляхом віддзеркалення зображення відносно його країїв. Просте копіювання межевих пікселів недопустиме, адже воно призведе до гістограм із яскраво вираженими піками.

## Властивості АВГ

----

- Розмір околу є параметром методу. Він слугує характерною одиницею довжиною: контраст у менших регіонах покращується, а контраст у більших регіонах зменшується.

- Через природу алгоритму вирівнювання гістограми, значення результату пікселя під дією АВГ пропорційне його рангу серед пікселів у його околі. Це дозволяє ефективно реалізовувати алгоритм на спеціалізованому обладнанні, яке може порівнювати центральний піксель із усіма іншими пікселями в околі.<sup>[3]</sup> Не нормалізоване значення результату можна обчислити, додавши $2$ для кожного пікселя з меншим значенням, ніж центральний піксель, і додавши $1$ для кожного пікселя з тим же значенням.

- Коли окіл якогось із пікселів складається із пікселів із майже однаковими інтенсивностями, то його гістограма матиме яскраво виражений пік, і перетворення переведе відносно вузький проміжок інтенсивностей на увесь проміжок $[0, 255]$ у результуючому зображенні. Це означає, що АВГ перебільшує малий шум у здебільшого однорідних областях зображення.<sup>[4]</sup>

## Контрастно-обмежене АВГ

-----

Звичайне АВГ має схильність перебільшувати шук у здебільшого однорідних областях зображення, оскільки гістограма у таких областях дуже концентрована. Як наслідок, АВГ може спричиняти збільшення шуму у майже однорідних областях. Контрастно-обмежене АВГ (КОАВГ) є варіантом адаптивного вирівнювання гістограми, у якому збільшення контрасту є обмеженим, задля зменшення проблеми зашумлення.<sup>[3]</sup>

У КОАВГ підсилення контрасту в околі заданого значення пікселя задається похідною функції перетворення. Вона є пропорційною похідній [функції розподілу](https://uk.wikipedia.org/wiki/%D0%A4%D1%83%D0%BD%D0%BA%D1%86%D1%96%D1%8F_%D1%80%D0%BE%D0%B7%D0%BF%D0%BE%D0%B4%D1%96%D0%BB%D1%83_%D0%B9%D0%BC%D0%BE%D0%B2%D1%96%D1%80%D0%BD%D0%BE%D1%81%D1%82%D0%B5%D0%B9) околу, а отже і значенню гістограми при цьому значенні інтенсивності пікселя. КОАВГ обмежує підсилення контрасту шляхом відсікається гістограми на заздалегідь заданому рівні, яке робиться перед обчисленням ФР. Це обмежує похідну ФР, а отже і похідну функції перетворення. Значення, при якому обрізається гістограма, так звана межа відтину, залежить від нормалізації гістограми, а отже і від розміру околу. На практиці зустрічаються межі між $3$ і $4$.

Бажано не відкидати частину гістограми, яка перевищує межу відтину, а перерозподіляти її рівномірно між усіма кошиками гістограмами.<sup>[3]</sup>

![Clahe-redist](https://upload.wikimedia.org/wikipedia/commons/thumb/5/5f/Clahe-redist.svg/300px-Clahe-redist.svg.png)

Перерозподіл знову підніме деякі області над межею відтину (область, зафарбована зеленим кольором на рисунку). Таким чином отримується ефективна межа відтину, яка перевищує встановлений ліміт, і точне значення якої залежить від зображення. Якщо це небажано, процедуру перерозподілу можна повторювати рекурсивно, поки надлишком не можна буде знехтувати.

## Ефективне обчислення шляхом інтерполяції

----

Адаптивне вирівнювання гістограми, у тому прямолінійному вигляді, у якому воно представлене вище, як з так і без обмеження контрасту, вимагає обчислення **різних** гістограм та перетворювальних функцій для кожного із пікселів. Це робить метод дуже затратним з обчислювальної точки зору.

Інтерполяція дозволяє значно підвищити ефективність без шкоди для якості результату.<sup>[3]</sup> Зображення розділяється на прямокутні блоки однакового розміру, як показано у правій частині малюнка нижче. (На практиці частим вибором є 64 блоки в 8 стовпцях та 8 рядках.<sup>[4]</sup>) Потім для кожного з блоків обчислюється гістограма, функція розподілу та перетворювальна функція. Отримані функції одразу є правильними для центрів блоків, чорних квадратів у лівій частині фігури. Усі інші пікселі перетворюються за допомогою лінійної комбінації перетворювальних функцій, що відповідають чотирьом найближчим блокам, з відповідними ваговими коефіцієнтами. Пікселі всередині зображення (заштриховані синім) [інтерполються білінійно](https://uk.wikipedia.org/wiki/%D0%91%D1%96%D0%BB%D1%96%D0%BD%D1%96%D0%B9%D0%BD%D0%B0_%D1%96%D0%BD%D1%82%D0%B5%D1%80%D0%BF%D0%BE%D0%BB%D1%8F%D1%86%D1%96%D1%8F), пікселі поблизу межі (заштриховані зеленим) [інтерполються лінійно](https://uk.wikipedia.org/wiki/%D0%9B%D1%96%D0%BD%D1%96%D0%B9%D0%BD%D0%B0_%D1%96%D0%BD%D1%82%D0%B5%D1%80%D0%BF%D0%BE%D0%BB%D1%8F%D1%86%D1%96%D1%8F), а пікселі поблизу кутів (заштриховані червоним) перетворюються за допомогою перетворювальної функції кутового блоку. Коефіцієнти інтерполяції відображають розташування пікселя між найближчими до нього центрами блоків, і підібрані таким чином, щоб результат був неперервним у центрах блоків.

![Clahe-tileinterpol](https://upload.wikimedia.org/wikipedia/commons/thumb/4/49/Clahe-tileinterpol.svg/450px-Clahe-tileinterpol.svg.png)

Ця процедура драматично зменшує кількість перетворювальних функцій перетворення, які необхідно обчислити, і вимагає лише невеликі додаткові витрати на лінійну інтерполяцію на заключному етапі алгоритму.

## Ефективне обчислення шляхом інкрементальних оновлень гістограми

----

Альтернативою до замощення зображення є ідея рухати вікно (окіл) на один піксель за раз, і оновлювати гістограму,<sup>[5]</sup> додаючи новий ряд, і віднімаючи залишений позаду ряд. Цей алгоритм був названий SWAHE (Sliding Window Adaptive Histogram Equalization) його авторами. Обчислювальна складність обчислення гістограми таким чином зменшується з $O(N^2)$ до $O(N)$ (де $N$ --- розмір сторони вікна у пікселях). Цей алгоритм також не вимагає заключного кроку інтерполяції.
